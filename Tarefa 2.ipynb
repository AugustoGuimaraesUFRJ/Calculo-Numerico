```julia
#Exercício 2.1
#Função para achar a raiz cúbica de 43
function raiz_cubica_43(chute, iteracoes)
    for i=1: iteracoes
        chute = chute - (chute*chute*chute - 43)/(3*chute*chute) #"andando" pela reta tangente
    end
    return chute #aproximação
end
```

```julia
raiz_cubica_43(3,1)
```

```julia
#Exercício 4.5
#Utilizando o video do professor Abel como base, foi possivel montar o seguinte algoritmo
#Função que calculo a serie de Taylor de ln(x) na ordem n
function ln(x,n)
    # 0 + (x-1) - (x-1)**2/2 + 2(x-1)**3/6 ...
    s = 0.0
    for k=1: n
        s += (((-1)^(k-1)) * ((x-1)^k))/k
    end 
    return s #Aproximação
end
```

```julia
ln(1.5, 3)
```

```julia
#Utilizando o video do professor Abel como base, foi possivel montar o seguinte algoritmo
#Função que calculo a serie de Taylor de ln(x) para um valor x com um erro máximo
function lnErro(x,Erro)
    s = 0.0
    k= 1
    #Enquanto a diferença entre o valor de ln(x) e a aproximação for maior que Erro dado, ele vai executar o metodo 
    while(abs(log(x) - s) >= Erro) 
        s += (((-1)^(k-1)) * ((x-1)^k))/k
        k+=1
    end 
    return s #Aproximação
end
```

```julia
lnErro(1.5,0.01)
```

```julia
#Exercício 4.7
#Utilizando a formula cos(x) = Somatorio (-1)^n * (x^(2n))/(2n!) Serie de Taylor aprendida em Calculo 4
#https://www.youtube.com/watch?v=bowHvSlam7M
using Plots
gr(size=(600,400))
f(x)=cos(x)
g(x)= 1.0
h(x)=1-x*x/2
k(x)=1-x*x/2 + x^4/(2*3*4)
l(x)=1-x*x/2 + x^4/(2*3*4) - x^6/(2*3*4*5*6)
p(x)=1-x*x/2 + x^4/(2*3*4) - x^6/(2*3*4*5*6) + x^8/(2*3*4*5*6*7*8)
plot(f,-6,6,label="cos()function", c=:blue,
    grid=false,
    axis=true,
    xaxis=([-6.5,6.5]),
    xticks=([-6,-4,-2,0,2,4,6]),
    yaxis=([-6.5,4]),
    yticks=([-6,-4,-2,0,2,4]))
plot!(g,-6,6,label="Taylor series - 1 terms", c=:orange)
plot!(h,-6,6,label="Taylor series - 2 terms", c=:green)
plot!(k,-6,6,label="Taylor series - 3 terms", c=:red)
plot!(l,-6,6,label="Taylor series - 4 terms", c=:purple)
plot!(p,-6,6,label="Taylor series - 5 terms", c=:brown)
plot!(legend=:bottomleft)
```
